### urllib.parse.urlencode() 함수로 Query 문자열(요청 파라미터) 인코딩하기

```python
import urllib
print("[ Query문자열 또는 요청 파라미터 인코딩 ]")
params1 = urllib.parse.urlencode({'number': 12524, 'type': 'issue', 'action': 'show'})
print(params1)
params2 = urllib.parse.urlencode({'addr': '서울시 강남구 역삼동'})
print(params2)
```

### urllib.parse.quote_plus()는 urlencode와 달리 value값만 전달한다. 둘이 목적은 같은 함수다. name:value를 같이 인코딩하면 urlencode

```python
import urllib
print(urllib.parse.quote_plus('가나다 ABC123'))
query_str1 = "q="+urllib.parse.quote_plus('가나 다')
query_str2 = urllib.parse.urlencode({"q" : "가나 다"})
print("-------------------------------------------")
print(query_str1)
print(query_str2)
```

### Query 문자열 전달하면서 GET 방식으로 요청하기 (?뒤에 %s는 파이썬 구문으로 %s는 문자열이 온다는 뜻이다. 이 자리에 %params가 온다.)

```python
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({'name': '유니코', 'age': 10})
print("URL 인코딩 규칙이 적용된 문자열 : %s" % params)
url = "http://unico2013.dothome.co.kr/crawling/get.php?%s" % params
print("URL 문자열 : %s" % url)
with urllib.request.urlopen(url) as f:
    print(f.read().decode('utf-8'))
```

### 요청 파라미터 전달하면서 POST 방식으로 요청하기(1)

```python
import urllib.request
import urllib.parse
data = urllib.parse.urlencode({'name': '유니코', 'age': 10})
print("URL 인코딩 규칙이 적용된 문자열 : %s" % data)
postdata = data.encode('ascii')
print("변환된 바이트 문자열 : %s" % postdata)
url = "http://unico2013.dothome.co.kr/crawling/post.php"
with urllib.request.urlopen(url, postdata) as f:
    print(f.read().decode('utf-8'))
```

### 요청 파라미터 전달하면서 POST 방식으로 요청하기(2)

```python
import urllib.request
import urllib.parse
data = urllib.parse.urlencode({'name': '유니코', 'age': 10})
postdata = data.encode('ascii')
req = urllib.request.Request(url='http://unico2013.dothome.co.kr/crawling/post.php',
            data=postdata)
print(req)
with urllib.request.urlopen(req) as f: #with 블록을 as 변수(f)에 저장하겠다는 뜻
    print(f.read().decode('utf-8'))
```

### requests 패키지는 urllib와 다르게 값을 넘겨줄때 튜플이든 리스트든 딕셔너리든 간에 인코딩을 해주지 않아도 된다.(서버한테 쿼리문자열을 보낼 때 urllib는 urlencode()로 보낼 것들을 인코딩하고 GET방식의 경우 그걸 직접 뒤에 '%'붙여서 해야하는데 requests는 그런 게 필요가 없다.)

```python
import requests

dicdata = {'key1': 'value1', 'key2': 'value2'}
urlstr = 'http://unico2013.dothome.co.kr/crawling/get.php'
r = requests.get(urlstr, params=dicdata)
print(r.url)
print('------------------------------------')
dicdata = {'key1': 'value1', 'key2': ['value2', 'value3']}
r = requests.request('GET', urlstr, params=dicdata)
print(r.url)
print('------------------------------------')
tupledata = [('key1', 'value1'), ('key1', 'value2')]
r = requests.get(urlstr, params=tupledata)
print(r.url)
```

### 밑에 코드는 영화 리뷰 글 읽어오기 코드다.

```python
import requests
from bs4 import BeautifulSoup
import re
req = requests.get('http://movie.naver.com/movie/point/af/list.nhn?page=1')
html = req.text
soup = BeautifulSoup(html, 'html.parser') #print(html)을 했을때 글자가 안 깨지면 text로 받아올때 encoding안해줘도 BeautifulSoup이 알아서 인코딩 해준다.
titles = soup.select('.movie')
points = soup.select('td.title > div > em')
reviews = soup.select('td.title')
movie_title = []
movie_point = []
movie_review = [] 

for dom in titles:
    movie_title.append(dom.text)
for dom in points:
    movie_point.append(dom.text)
for dom in reviews:
    content = dom.contents[6] #0부터 시작해서 6 즉 7번째 #contents는 자식을 리스트객체로 다 꺼내준다. 텍스트가 됐든 태그가 됐든(엔터키 포함) 홈페이지에서 코드를 보면 DOM객체가 7번째일때 리뷰글이 나온다.
    content=re.sub("[\n\t]", '', content) #import re 필요 #R에서의 gsub와 마찬가지로 대체하는 역할을 하는 함수다.   
    movie_review.append(content)
commentLength = len(movie_title)   
for i in range(commentLength):
    print("영화 제목 : " + movie_title[i])
    print("평점 : " + movie_point[i])
    print("리뷰글 : " + movie_review[i])
    print("-----------------------------------------")   
```

```python
import requests
from bs4 import BeautifulSoup
import re
for n in range(1,31):
    req = requests.get('http://movie.naver.com/movie/point/af/list.nhn?page='+str(n))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser') #print(html)를 했을때 글자가 안 깨지면 text로 받아올때 encoding안해줘도 BeautifulSoup이 알아서 인코딩 해준다.
    titles = soup.select('.movie' )
    points = soup.select('td.title > div > em')
    reviews = soup.select('td.title')
    movie_title = []
    movie_point = []
    movie_review = [] 
    for dom in titles:
        movie_title.append(dom.text)
    for dom in points:
        movie_point.append(dom.text)
    for dom in reviews:
        content = dom.contents[6]
        content=re.sub("[\n\t]", '', content)
        movie_review.append(content)

    commentLength = len(movie_title)   
    for i in range(commentLength):
        print(movie_point[i] + ","+movie_title[i]+","+movie_review[i])
    print("-----------------------------------------------------")
```

```python
import requests
from bs4 import BeautifulSoup
title = []
link = []
urlstr = 'http://www.yes24.com/SearchCorner/Search?domain=BOOK&query=python'
r = requests.get(urlstr)
#r.encoding = "utf-8" #한글 처리를 안 해도 BeautifulSoup한테 주면 알아서 처리해준다. #print(r.text)를 했을때 글자가 안 깨지면 text로 받아올때 encoding안해줘도 BeautifulSoup이 알아서 인코딩 해준다.
bs = BeautifulSoup(r.text, 'html.parser')
bookList = bs.select('a.gd_name') #a태그의 gd_name 클래스

for bookDom in bookList:
    title.append(bookDom.string) #a태그의 컨텐트를 꺼내서 붙인다.
    link.append(bookDom["href"]) #a태그의 href 속성의 속성값을 꺼내서 붙인다.

print("-- 도서 제목 --")
print(title)
print("-- 도서 링크 URL --")
print(link)
with open('output/booklink.csv', "wt", encoding="utf-8") as f: #output 폴더를 안 만들었으면 에러가 난다. #close()를 해줘야 하는데 with 블록에 해서 생략 가능하다고 함.
    f.write('booktitle,booklink\n')  
    for i in range(len(title)):
        f.write(title[i]+","+link[i]+'\n')  
```

##### 위에 코드에서 output/booklink.csv를 열어서 보면 '일잘러의 비밀, 엑셀 대신 파이썬으로 업무 자동화하기'안에 ','가 들어가있어서 문제가 발생함을 볼 수 있다. 이럴때는 csv에 넣기전에 ','를 빼줘야 한다. 이런 문제가 발생한 이유는 저장은 제대로 됐는데 파일을 읽을때 열을 구분하는 단위가 쉼표(',')라서 그렇다. 쉼표가 있으면 다음 열로 인식한다. csv파일을 직접 메모장으로 열어서 보자.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


titles = list()
summaries = list()
ratings = list()

for page_num in range(1,21):
    url = f"https://comic.naver.com/genre/bestChallenge.nhn?&page={page_num}" 
    req = requests.get(url)
    
    html = req.content.decode("utf-8") 
    bs= BeautifulSoup(html, "html.parser") #BeautifulSoup 객체 생성
    
    for info in bs.find_all(class_="challengeInfo"): #코드를 열어보면 알겠지만 challengeInfo라는 클래스 속성을 가진 것이 만화 하나에 대한 정보다. #find_all()을 하면 ResultSet 객체가 리턴된다. 그걸로 for문을 돌린다.
        titles.append(info.find(class_="challengeTitle").a.text.strip()) #태그 객체.find를 하면 해당 객체의 자손에서 찾는다. BeautifulSoup객체.find하면 그 객체에서 찾는거고. 즉 이건 challengeInfo 클래스를 가진 태그의 자손에서 challengeTitle 클래스를 갖는 태그를 찾는 것이다.
        summaries.append(info.find(class_="summary").text.strip())
        ratings.append(info.find(class_="rating_type").strong.text.strip())
    print(f"{len(titles)}개 추출")
    
#이 코드를 3개로 나눴다. 아래 셀 2개도 이 코드와 하나다.
```

```python
naver_comic_dict = { 'title':titles, 'summary' : summaries, 'rating' : ratings }    #딕셔너리 객체를 만든다. 각 key에 대해 value들은 480개의 데이터를 가지고 있다. 'title', 'summary', 'rating'들이 변수(column명)가 된다.
print(naver_comic_dict)
```

```python
df = pd.DataFrame(naver_comic_dict) #데이터 프레임 생성 #R에서는 data.frame이었다. #pd는 pandas. #데이터프레임으로 만들고싶은 딕셔너리를 매개변수로 전달함.
display(df) #데이터 프레임 일부 출력 #print보다는 display가 훨씬 보기 좋다. visualtic하기 때문. #위에서 5개, 밑에서 5개를 보여준다. 판다스는 행넘버를 인덱스라고 부른다.
df.to_csv("output/webtoon.csv") #데이터프레임을 csv 파일로 저장한다.
```

### API 별 응답내용 한글처리 방법

```python
import requests
from bs4 import BeautifulSoup
urlstr = 'http://unico2013.dothome.co.kr/crawling/get.php'
r = requests.get(urlstr)
r.encoding = 'utf-8' #이 경우에는 해서 줘야한다. text로 꺼낼때는 한글을 utf-8로 인식해서 문자로 변환하는게 아니라 윈도우 기본 운영체제의 기본 charset으로 해서 한글이 깨지는 것이라고 처음에 강사가 설명했는데 내가 실험해보니 아닌 것 같아 질문을 했고 아래와 같은 답변을 받았다.
html = r.text #encoding이나 decode는 text나 content로 받아오기 전에 하거나 받아오고나서 하거나 둘 다 할 수 있다. 둘의 차이가 뭔지 생각해보자. 지금은 글로 적을만큼은 이해 안된다.
print(html)     #다만 encoding을 안 해줘도 print(html)했을 때 글자가 안 깨졌다면 인코딩을 안 해주고 BeautifulSoup 객체로 출력해도 글자가 깨지지 않는다. 여기서는 인코딩을 하지 않고 print(html)을 하면 글자가 깨져서 인코딩을 해줘야 한다.
print("===================================") #이 코드에서 인코딩을 안 했을때 print(html)을 하면 글자가 깨지는 이유는 강사가 위 사이트를 utf-8(BOM)으로 작성했기때문일 것이라고 한다.(다른 사이트들은 text든 content든 인코딩이나 디코드 안 해도 다 BeautifulSoup 객체가 알아서 처리해준다.) 
bs = BeautifulSoup(html, 'html.parser')
print(bs.prettify()) #BeautifulSoup 객체로 prettify()로 보기 좋게 출력
```

```python
import requests
from bs4 import BeautifulSoup
urlstr = 'http://unico2013.dothome.co.kr/crawling/get.php'
r = requests.get(urlstr)
html = r.content
html = r.content.decode('utf-8') #바이트열로 오면 BeautifulSoup 객체는 바이트열을 decode 처리를 알아서 해준다. 즉 이 라인을 주석처리해도 맨 마지막 라인은 정상적으로 한글을 출력해준다.
print(html)
print("===================================")
bs = BeautifulSoup(html, 'html.parser')
print(bs.prettify())
```
