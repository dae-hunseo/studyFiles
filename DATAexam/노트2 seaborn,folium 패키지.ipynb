### seaborn 스타일링
- set_style: background color, grid, spine, tick을 정의하여 그림의 전반적인 모양을 스타일링
- set_context: 프리젠테이션이나 보고서와 같은 다양한 매체에 활용할 수 있도록 스타일링

1. set_style : 그림의 전반적인 모양 스타일링
    - Built-in Themes (내장 테마) 활용하기
    - darkgrid, whitegrid, dark, white, ticks.
    - 기본값은 darkgrid이지만, 원하는대로 변경이 가능
2. sns.despine() 함수
    - 축선 표시 여부 결정하는 함수(right=T/F, left=T/F, top=T/F, bottom=T/F)
        - 각 속성값이
        - True 면: 표시하지 않는다.
        - False 면: 표시함

```python
#seaborn이용한 style 설정

x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

#sns.set_style('whitegrid')      #흰색 배경에 grid가 추가된 스타일
#sns.set_style('darkgrid') #회색 배경에 grid가 추가된 스타일
sns.set_style('dark') #그리드 없이 배경이 회색

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
#seaborn이용한 style 설정
x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

sns.set_style('dark')             #배경 회색 # 한 번설정하면 바꾸지 않으면 그 다음 셀들에도 그대로 유지된다.
plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
#sns.despine() 함수
#축선 표시 여부 결정하는 함수(right=T/F,left=T/F,top=T/F,bottom=T/F)
#각 속성값이 True 면 : 표시하지 않는다.   #반대로 표현한다고 생각하면 안 됨.
#            False 면: 표시함            #반대로 표현하지 않는다고 생각하면 안 됨.
x=np.linspace(1,14,100)
y1=np.sin(x)
y2=2*np.sin(x)
y3=3*np.sin(x)
y4=4*np.sin(x)

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
plt.plot(x,y1,x,y2,x,y3,x,y4)
#sns.despine(right=True, left=True, top=False, bottom=False) #위와 아래쪽만 축선을 표시한다.
#sns.despine(right=False, left=False) #top은 별 다른 말을 안 했더니 표현이 안 됐다. #bottom은 별 다른 말을 안 했더니 표현이 됐다.
      #top=True, bottom=False가 기본값이다.

sns.despine(right=True, left=True) #bottom=False가 기본값인 것을 알 수 있음.
plt.show()
```

###  Distribution Plot
- Distribution Plot은 데이터의 분포를 시각화하는데 도움이 됨
- 이 그래프를 사용하여 데이터의 평균(mean), 중위수(median), 범위(range), 분산(variance), 편차(deviation) 등을 이해 할 수 있음

1. Hist Plot
- 변수에 대한 히스토그램을 표시한다.
- 하나 혹은 두 개의 변수 분포를 나타내는 전형적인 시각화 도구로 범위에 포함되는 관측수를 세어 표시한다.

```python
sns.histplot(x=tips['total_bill']) #total_bill을 구간으로 나누고 그 구간안에 데이터가 몇 개나 들어오는지 빈도센다.
```

2. KDE Plot
- 하나 혹은 두 개의 변수에 대한 분포를 그림
- histplot은 절대량이라면 kdeplot은 밀도 추정치를 시각화(히스토그램 막대의 중점들을 연결해서 곡선으로 그래프를 그려놓은 것이다.)
    - 결과물로는 연속된 곡선의 그래프를 얻을 수 있음

```python
sns.kdeplot(x=tips['total_bill'])
```

3. Box Plot
- 최대(maximum), 최소(minimum), mean(평균), 1 사분위수(first quartile), 3 사분위수(third quartile)를 보기 위한 그래프
- 특이치(outliar)를 발견하기에도 좋다.
- 단일 연속형 변수에 대해 수치를 표시하거나, 연속형 변수를 기반으로 서로 다른 범주형 변수를 분석할 수 있다.

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기

sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.boxplot(x=tips['total_bill'])
sns.despine()
plt.show() #10이상 40이하가 정상이고 40이상은 이상치로 표현하고 있다. 이런 이상치들은 성능을 높이는 방법 중에 하나로서 보통 제거하고 넣는다.
```

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기
#요일별 영수증 금액의 분포 확인

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

sns.boxplot(x=tips['day'], y=tips['total_bill']) #이러면 x값인 요일별로 카테고리를 나누고 박스  플롯을 그리게 된다.
sns.despine(offset=10) #offset: 축 간격 설정(10 정도 떨어트려라)
plt.show() #범주형 데이터별로 연속형 데이터를 그렸다 #주말로 갈수록 더 많은 금액을 지불하는 사람들이 늘어난다는 것을 그래프를 통해 알 수 있다.
```

```python
#tips 데이터를 이용한 box plot 그리기
#seaborn 패키지의 boxplot() 이용해서 그리기
#요일별 영수증 금액의 분포 확인
#palette 인수 사용 예제 : palette = '파레트명'

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')#x와 y에 tips[]를 안 써도 되도록 data로 지정해준다. #palette는 색상 지정으로 Set1,2,3 이정도는 내장속성으로 있다.
sns.despine(offset=10) #offset: 축 간격 설정
plt.show()
```

```python
#hue 인수 : 카테고리값을 가지는 변수의 이름 지정하면
#카테고리에 따라 그룹을 만들어서 표시

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

#요일 카테고리에 대하여 성별 그룹을 생성
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3')#흡연 여부에 따라 그룹을 생성해 요일별로 흡연자와 비흡연자들이 낸 영수금액의 분포를 보여준다.
sns.despine(offset=10) #offset: 축 간격 설정
plt.show()
```

```python
#hue 인수 : 카테고리값을 가지는 변수의 이름 지정하면
#카테고리에 따라 그룹을 만들어서 표시

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='Set3')#흡연 여부에 따라 그룹을 생성해 요일별로 흡연자와 비흡연자들이 낸 영수금액의 분포를 보여준다.
sns.despine(offset=10) #offset: 축 간격 설정
plt.show()
```

# seaborn.lmplot() 
- 스캐터그래프를 그릴때 회귀선을 기본으로 출력해주는 그래프 함수
- size 인수 : 그래프 크기
    - height 인수로 변경해서 사용하라는 메시지가 출력될 수 있음  
- hue 인수 : 카테고리 데이터를 분류하여 그룹화하되 하나의 그래프에 두개의 값을 모두 표현
- col 인수 : 카테고리 데이터를 분류하여 그룹화하되 그룹의 개수만큼 그래프를 생성해서 표현한다.


```python
#dataset : tips 사용
sns.set_style('darkgrid')

#plt.scatter(x=tips['total_bill'], y=tips['tip']) #여기선 tips[]를 써서 data=tips를 안 썼음.
sns.lmplot(x='total_bill', y='tip', data=tips)

plt.show()
```

```python
#dataset : tips 사용
#회귀선 생략 코드 : 인수 :  fit_reg=False
sns.set_style('darkgrid')

sns.lmplot(x='total_bill', y='tip', data=tips,fit_reg=False)

plt.show()
```

```python
#hue는 그룹을 만듬
sns.set_style('darkgrid')

sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker')

plt.show()
```

```python
#영수증 금액과 팁의 관계성을 lmplot()으로 표현하되, 
#성별을 구분해서 표현하시오.
sns.set_style('darkgrid')

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')

plt.show()

```

```python
#영수증 금액과 팁의 관계성을 lmplot()으로 표현하되, 
#성별을 구분해서 표현하되 그래프를 각각 그리시오.
#hue 인수 : 카테고리 데이터를 분류하여 그룹화하되 하나의 
#그래프에 두개의 값을 모두 표현
#col 인수 : 카테고리 데이터를 분류하여 그룹화하되 그룹의  
#개수만큼 그래프를 생성해서 표현한다.
sns.set_style('darkgrid')

sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')

plt.show()

```
