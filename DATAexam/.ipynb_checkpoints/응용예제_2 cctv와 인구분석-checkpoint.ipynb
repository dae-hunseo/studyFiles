```python
# 서울시 인구현황 가져오기 - 엑셀파일 읽어오기
# pd.read_excel('파일명')
pop_seoul = pd.read_excel('./data/01.population_in_Seoul.xls') 
pop_seoul.head() #1행과 2행은 스킾해야함(0행은 컬럼명으로 따로 빠진다.).
pop_seoul.tail() #그리고 다 필요한게 아니라 일부 컬럼만 필요함. 자치구에 해당하는 컬럼이 있어야 하고 그 다음에 전체인구 합계가 필요하고 고령자에 대한 합계와 외국인데이터만 있으면 된다.#
```

```python
#필요 data 만 가져오기, 자치구, 전체인구수,한국인,외국인,고령자(B,D,G,J,N)
# usecols='선택열'
#2행을 skip하고 3행을 header 처리
pop_seoul = pd.read_excel('./data/01.population_in_Seoul.xls',
                          header=2,usecols = 'B,D,G,J,N') #header로 위에 2행은 헤더로 취급해서 안 가져온다는 것 같다. #엑셀에 열 인덱스가 알파벳으로 되어있음. usecols로 필요한 열만 추출함.
pop_seoul.head()
```

```python
#컬럼명 변경
#0: 구별  -> cctv 데이터프레임의 구별 컬럼과 동일한 컬럼명을 사용(두 df를 하나로 합칠때 키 컬럼으로 사용)
#1:인구수
#2:한국인
#3:외국인
#4:고령자
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                          pop_seoul.columns[1]:'인구수',
                          pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자'},inplace=True)
pop_seoul.head()
```

#### 가공필드 생성

```python
cctv_seoul.head()
```

```python
# cctv 데이터
# 설치대수가 적은 5개 구를 확인
# 정렬기능 이용
cctv_seoul.sort_values(by='소계').head()
#도봉구의 설치 대수가 가장 적다.
```

```python
# 설치대수가 많은 5개 구를 확인
cctv_seoul.sort_values(by='소계', ascending=False).head() #내림차순 정렬
#강남구의 설치 대수가 가장 많다.
```

```python
# Data상 최근 3개년동안 cctv가 얼마나 증가했는지 확인
# 각 연도별 cctv 대수는 해당년도에 설치한 대수를 의미함
# 최근 증가율 = (2014년+2015년+2016년)/2013년도이전 * 100
#데이터 컬럼 즉 시리즈객체들에 대해서 사칙연산을 진행하면 벡터화 연산을 해서 서로 다른 열들의 같은 행들끼리의 데이터끼리 계산을 한다. 여기서 벡터화 연산을 하려면 ()로 묶어줘야 하는 것 같다. 아니면 에러난다.
cctv_seoul['최근증가율'] = (cctv_seoul['2014년'] + cctv_seoul['2015년'] + cctv_seoul['2016년'])/cctv_seoul['2013년도 이전']*100 
```

```python
cctv_seoul.head() #'최근증가율'열이 새로 생긴것을 알 수 있다
```

```python
# Data 상 최근 3개년동안  cctv 증가율이 높은 3개 구를 확인
cctv_seoul.sort_values(by='최근증가율',ascending=False).head(3) 
```

#### 인구데이터

```python
# 어느구의 인구가 가장 많은가?
pop_Seoul.head()
pop_seoul.sort_values(by='인구수', ascending=False).head()
```

```python
# 고령자수가 많은 구
pop_seoul.sort_values(by='고령자', ascending=False).head()
# 외국인이 많은 구
pop_seoul.sort_values(by='외국인', ascending=False).head()
```

```python
# 인구데이터 가공필드 생성
# 전체 인구수 대비 외국인 비율 계산 후 컬럼 생성
pop_seoul['외국인비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100 #비율이라 100을 곱함
pop_seoul.head(2)
```

```python
# 전체 인구수 대비 고령자 비율 계산 후 컬럼 생성
pop_seoul['고령자비율']=pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head(2)
```

```python
pop_seoul.sort_values(by='고령자비율', ascending=False).head() #고령자비율이 높은 순(내림차순)
pop_seoul.sort_values(by='외국인비율', ascending=False).head() #외국인비율이 높은 순(내림차순)
```

```python
pop_seoul.columns
cctv_seoul.columns
```

```python
# 인구 대비 cctv 대수의 적정성을 확인해야 하므로 
# 두 DF를 병합 - 양쪽 df에 '구별'컬럼(동일한 의미)
# merge(data1, data2, on=기준컬럼)
data_result = pd.merge(cctv_seoul, pop_seoul, on='구별')
data_result.head()
```

```python
# 불필요한 열 삭제 
# cctv데이터의 2013-2016필드 삭제
del data_result['2013년도 이전'] #'2013년도 이전'열삭제
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
```

```python
data_result.head()
```

```python
#시각화 등에 편리하게 사용하기 위해 기준열인 '구별' 컬럼을 행인덱스로 처리
# df.set_index(컬럼명,[inplace=True]) #inplace=True를 해야 원본에 반영됨. []은 선택이란 뜻
data_result.set_index('구별', inplace=True)
```

```python
data_result.head()
```

```python
# 다수의 데이터 중에서 상관관계가 가장 큰 데이터를 비교
# 인구와 관련된 각 필드와 cctv 소계와의 상관계수를 파악해서 그래프로 표현

# 상관계수 : 두 변수의 관련성을 확인하는 방법
# 절대값이 클수록 두 데이터는 관계가 있다고 봄

# 절대값 기준
# 0.1이하면 무시
# 0.3이하면 약한 상관 관계
#0.7을 기준으로 강한 상관관계가 있다 봄 #0.8이 넘어가면 자기상관관계가 아닌지 확인할 필요가 있음. 자기 상관은 제거해야할 데이터중에 하나

# 상관관계 파악할 컬럼
# 절대적인 cctv 대수 : 인구수/고령자비율/외국인비율

# np.corrcoef(데이터값 1, 데이터값 2)
np.corrcoef(data_result['소계'], data_result['인구수'])
# 상관계수 : 0.2432198 - 약한 양의 상관
```

```python
np.corrcoef(data_result['소계'], data_result['고령자비율'])
# 상관계수 : -0.25866272 - 약한 음의 상관
```

```python
np.corrcoef(data_result['소계'], data_result['외국인비율'])
# 상관계수 : -0.05230165 - 절대값이 0.1이하이므로 두 변수는 상관성이 없다
```
