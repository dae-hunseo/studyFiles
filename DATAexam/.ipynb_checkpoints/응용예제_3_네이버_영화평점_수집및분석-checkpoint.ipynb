## 함수 구성
- 특정날짜를 파라미터로 전달받아서 영화명과 평점을 추출한 후에 데이터프레임으로 반환하는 함수
- get_movie(date):
    - date인수를 사용해 url 설정
    - date별로 영화정보(title, point)를 반환

```python
def get_movie(date):
    #URL 생성
    url_base = 'http://movie.naver.com/'
    url_sub = '/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'
    #문자열에 {변수명}을 사용하면 {}내부의 문자는 변수로 사용됨.(위의 {date}처럼) {}안은 아직 결정되지 않은 문자가 들어올 공간(변수) |#{}안에 들어올 내용은 나중에 format함수 이용해서 결정함.
    url = urljoin(url_base, url_sub)

    today=date
    url_fin = url.format(date=quote(today.strftime('%Y%m%d')))
    
    # 페이지 요청
    html = urlopen(url_fin)

    # 파싱객체 생성
    soup = BeautifulSoup(html, 'html.parser')

    # 랭킹 날짜 생성
    movie_date = today
    # 영화제목 추출
    movie_name = [s.a.text for s in soup.findAll('div', {'class', 'tit5'})]
    # 영화 평점 추출
    movie_point = [s.text for s in soup.findAll('td', {'class':'point'})]
    
    # 데이터프레임으로 반환
    return pd.DataFrame({'날짜':movie_date,
                         '영화제목':movie_name,
                         '영화평점':movie_point})
```

```python
# 2022-07-18 영화 데이터 추출
get_movie(date[17])
```

```python
movie_df = pd.DataFrame()
for today in date:
    movie_df = pd.concat([movie_df, get_movie(today)], axis=0, ignore_index=True)
```

## 시각화 작업
- 영화별 평점의 합계를 계산(평점의 합이 높은 영화를 추출)
    - 추출된 영화에 대해서 평점의 추이를 확인할 수 있는 그래프 작성

```python
# 데이터 특성확인
movie_df.info()
```

```python
# 영화평점의 Dtype은 object(문자열)
# 그래프 작업과 합계를 계산해야하므로 수치 형태로 변화 : astype(데이터타입)
movie_df['영화평점'] = movie_df['영화평점'].astype(float)
movie_df.info()
```

```python
# 영화별 Point의 합: pibot_table(data_set,index = [기준열], aggfunc = 계산함수) #계산함수는 평균, 합 등
movie_unique = pd.pivot_table(movie_df, index=['영화제목'], aggfunc=np.sum)
```

```python
display(movie_unique.head())
display(movie_unique.tail())
```

```python
# 영화 평점의 합이 높은 순으로 정렬
movie_top = movie_unique.sort_values(by='영화평점',ascending=False) #내림차순(높은순으로 먼저 나옴)
```

```python
movie_top.head(50)
```

```python
# 탑건: 매버릭 영화에 대한 평점 변화 그래프 시각화
# 탑건: 매버릭 레코드 추출
# df.query(조건)
tmp = movie_df.query('영화제목==["탑건: 매버릭"]') #컬럼 인덱스명 == ["해당하는 데이터"]
```

```python
tmp
```

```python
#평점 변화 그래프
#시계열 데이터이므로 line그래프
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,8))
plt.plot(tmp['날짜'], tmp['영화평점'])
plt.grid()
plt.show()
```

```python
movie_df #이런 데이터프레임을 long형 데이터 프레임이라 부른다.
```

```python
# 데이터 재배치 - long->wide로
# 날짜별 영화의 point를 확인하기 위해 재구성
# pivot_table 이용
# index = 날짜, column=영화제목 values=영화평점
movie_wide = pd.pivot_table(movie_df,
                            index=['날짜'],
                            columns = ['영화제목'],
                            values=['영화평점']) #values는 멀티인덱스로 데이터의 값을 뜻함. 멀티인덱스는 출력했을때 표의 맨 위에 보면 덩그랗게 혼자 구석에 있음
```

```python
movie_wide.shape
```

```python
display(movie_wide.head())
display(movie_wide.tail())
```

```python
# 피봇테이블 구성 중에 멀티인덱스로 구성 되었음
# 최상위 인덱스(영화평점)이 불필요하므로 제거
# 멀티인덱스 제거
# df.columns.droplevel()
movie_wide.columns = movie_wide.columns.droplevel()
movie_wide.head()
```
