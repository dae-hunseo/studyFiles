- 데이터 형태 및 특성 확인

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

- 농어의 길이가 길어짐에 따라 무게도 늘어남
    - 농어 길이 특성을 이용해 무게를 예측하자...

- 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(perch_length,
                                                                     perch_weight,
                                                                     random_state=42) #random_state를 설정해서 랜덤으로 매번 같은 값이 나오게 함.
```

```python
print(train_input.shape, test_input.shape)
```

- 학습 데이터가 1차원임
    - 사이킷런은 2차원 입력을 요구하므로 차원을 변경해야 함
    - 넘파이 배열.reshape(행,열[,깊이]) 함수 사용

```python
# shape 속성 - 튜플구조를 반환
test_array = np.array([1,2,3,4])
test_array.shape[0]
```

```python
# reshape() :  데이터의 차원을 변경하는 함수
test_array = test_array.reshape(2,2) #행렬로 변경
print(test_array.shape) #2행 2열
```

```python
# 행렬로 변경했을 때 나타나는 원소수가 원본과 일치해야 함
# 2행3열은 6개의 원소, 원본은 4개 원소
# test_array = test_array.reshape(2, 3) #개행렬로 변경했을 때 나타나는 원소수가 원본과 개수가 다르면 에러가 남
```

- input 데이터 차원 변경 : 행에 -1을 사용하면 행은 알아서 계산
    - 행은알아서계산, 1열 의 2차원 데이터로 변환됨

```python
train_input = train_input.reshape(-1, 1) #-1은 알아서 하라는 뜻. 행은 자동, 열은 1행
test_input = test_input.reshape(-1, 1)
```

```python
print(train_input.shape, test_input.shape)
```

## 결정 계수 ($ R^2$) (1이 나왔으면 100% 성능을 가진 것이다. 0.77이 나왔다면 굉장히 잘 나온 것. 0.4, 0.5도 꽤 좋은 모델로 평가받음. 앞에서 나온 정확도와는 개념이 조금 다르다. 얘는 대략 얼마정도 된다고 예측을 하는 것이지, 반드시 정확히 얼마라고 예측을 하지 못하기 때문에 분류보다 훨씬 더 오차가 많은 예측이 회귀에 해당이 된다.)
- 회귀 모델의 예측 결과에 대한 평가지표
- 회귀에서 정확한 숫자(정확한 무게)를 맞춘다는 것은 <span style="color:red">불가능</span> 함
    - 예측값이나 타깃값 모두 임의의 숫자임
    - Ex. A지역의 B 아파트 25평의 가격을 생각해보면
        - 해당 가격은 그 아파트가 매매가 이루어질때 결정됨
        - 동일한 날 동일한 부동산에서 거래가 이루어져도 다른 가격에 거래가 이루어 질 수 있음
        - 그러므로 부동산 가격은 절대적인 가격이 아닌 임의의 값으로 봐야 함
    - 결국 회귀의 평가는
        - 모든 타겟에 대해서 타깃값에서 예측값을 뺀 크기가 얼마나 되는지를 확인해서 계산 하게됨
        

평균보다 얼마나 떨어져서 예측이 되냐를 비율로 나타낸 것이 R2이다.

```python
from sklearn.neighbors import KNeighborsRegressor #Regressor는 회귀에 사용한다는 뜻이다.
```

```python
knr = KNeighborsRegressor() #모델 객체를 하나 만든다.

# k-최근접 이웃 회귀 모델을 훈련합니다
knr.fit(train_input, train_target)
```

```python
knr.score(test_input, test_target) #회귀모델(KNN)의 기본 평가지표는 R2값임. 
#최댓값이 1인데 0.99가 나왔다. 이건 성능의 지표로 쓸 때는 좋다고 이야기할 수 있긴하다. 그런데 이게 농어의 무게에 얼마나 오차가 나는지 확인하고자 한다면 이 R2값 보단느 다른 지표를 이용하면 오차가 얼마나 나는지를 확인해볼 수 있다.
```

- $ R^2$ 지표로 0.99가 나옴
    - 상당히 좋은 값 이지만 직관적이지 못함
    - 직관적인 다른 지표를 사용
        - MAE : 타깃값과 예측값의 절대값 오차의 평균

만약
30인데 35로 예측하면 5가 오차가 나고
40인데 38로 예측하면 오차는 절대값이기 때문에 2가 나온다.
- mae : (5+2)/2

```python
from sklearn.metrics import mean_absolute_error
```

```python
# 테스트 세트에 대한 예측을 만듭니다
test_pred = knr.predict(test_input)

# 테스트 세트에 대한 평균 절댓값 오차를 계산합니다
mae = mean_absolute_error(test_target, test_pred)
mae #knr.score(test_input, test_target)에서 99%가 나왔기 때문에 무게 19g 정도 차이는 좋다고 보는 거다.
```

- 결과적으로 평균 19g정도 타깃값과 다르다는 의미

## 과대적합 vs 과소적합

### 테스트 보다 학습이 덜 나오면 과소적합이라 한다.

```python
print(knr.score(test_input, test_target))
```

```python
print(knr.score(train_input, train_target)) #첫째자리가 차이난 것도 아니기 때문에 3%도 안되는 이정도 차이는 과소적합으로 보지 않는다.
```

- 훈련 세트의 score보다 테스트 세트의 score가 더 높다
    - <span style="color:red">과소 적합</span>(차이가 이정도 미비한 것은 그냥 넘어간다)

```python
# 이웃의 갯수를 3으로 설정합니다
knr.n_neighbors = 3

# 모델을 다시 훈련합니다
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
```

```python
print(knr.score(test_input, test_target))
```

- 훈련 세트의 score가 테스트 세트의 score보다 더 높다
    - <span style="color:red">과대 적합</span>(차이가 이정도 미비한 것은 그냥 넘어간다)

- 차이가 미비한건 상관이 없음(실제 작은 차이는 과대/과소 적합이라고도 하지 않음)
    - 차이가 너무 많이 나면
        - 훈련을 너무 많이 시켜 훈련데이터에만 맞는 모델이 되었거나(과대적합)
            - 과대적합은 훈련스코어는 높으나, 테스트 스코어는 현저하게 낮은 경우
        - 훈련을 너무 적게 해서 제대로 된 결과를 내지 못하고 있는 것(과소적합)
            - 과소 적합은 성능이 굉장히 떨어진 결과로 나타남

## 과대/과소 적합 확인 예제
- n_neighbors 값을 변경시켜가면서 그래프를 통해 과대,과소 적합 확인

```python
# k-최근접 이웃 회귀 객체를 만듭니다
knr = KNeighborsRegressor()

# 5에서 45까지 x 좌표를 만듭니다
x = np.arange(5, 45).reshape(-1, 1)

# n = 1, 5, 10일 때 예측 결과를 그래프로 그립니다.
for n in [1, 5, 10]:
    knr.n_neighbors=n
    knr.fit(train_input, train_target)
    prediction = knr.predict(x)
    #훈련 세트와 예측결과 그래프 그리기
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.title('n_neighbors = {}'.format(n))    
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()
    #n_neighbors가 1인 경우는 과대적합, 5는 일반적이라고 볼 수 있다. 
    #10은 데이터들을 찾아가려고 노력하지 않고 그냥 직선을 그어버린 것과 같아진다. 오차가 너무 많아 이런 경우는 과소적합됐다고 부른다.
```
