```python
# 예제 3-3
import pandas as pd

# read_csv() 함수로 df 생성
df = pd.read_csv('./data/auto-mpg.csv', header=None)

# 열 이름을 지정
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

# 평균값 
print(df.mean(numeric_only=True))
print('\n')
print(df['mpg'].mean())
print(df.mpg.mean())
print('\n')
print(df[['mpg','weight']].mean())
```

```python
# 중간값 
print(df.median(numeric_only=True))
print('\n')
print(df['mpg'].median())
```

```python
# 최대값 
print(df.max(numeric_only=True))
print('\n')
print(df['mpg'].max())
```

```python
# 최소값 
print(df.min())
print('\n')
print(df['mpg'].min())
```

```python
# 표준편차 
print(df.std(numeric_only=True))
print('\n')
print(df['mpg'].std())
```

```python
# 상관계수  
print(df.corr())
print('\n')
print(df[['mpg','weight']].corr())
```

```python
# 예제 3-4
import pandas as pd

df = pd.read_excel('./data/남북한발전전력량.xlsx')  # 데이터프레임 변환 
display(df)
df_ns = df.iloc[[0, 5], 2:]            # 남한, 북한 발전량 합계 데이터만 추출 #2:는 3번째열부터 끝까지란 뜻 #0:5가 아닌 0,5다. header를 설정하지 않으면 제일 위에 행은 컬럼명으로 빠지니까 그 다음 0번째행, 5번째행만 추출하는 것이다.
display(df_ns)
df_ns.index = ['South','North']        # 행 인덱스 변경
display(df_ns)
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경 #df_ns는 6번째 라인에서 만든건데 문자열인 0번째, 1번째 column명은 제외되었음.
print(df_ns.head())
```

```python
def outliers(sr):
    Q1= sr.quantile(0.25)  # 1사분위 
    Q3 = sr.quantile(0.75) # 3사분위
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    return upper_limit, lower_limit
```

```python
outliers(sr)
```
