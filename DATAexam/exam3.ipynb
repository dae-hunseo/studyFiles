```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

url = "https://www.megabox.co.kr/movie-detail/comment?rpstMovieNo=21033000"
driver =  webdriver.Chrome(service=Service('C:/Temp/chromedriver'))
driver.get(url)

points = list()
recommends = list()
txts = list()

# 1페이지 정보
driver.implicitly_wait(3)
time.sleep(2)
points.extend([p.text for p in driver.find_elements(By.CLASS_NAME, 'story-point')]) #엘리먼트값 하나를 지정하여 추가할때는 append, 또 다른 리스트가 갖고있는 엘리먼트들을 엘리먼트로서 확장하고싶으면 extend. 만약 그냥 append를 쓰면 리스트가 붙는다.
recommends.extend([r.text for r in driver.find_elements(By.CLASS_NAME,'story-recommend')])
txts.extend([t.text for t in driver.find_elements(By.CLASS_NAME,'story-txt')])

# 2~30 페이지 정보
for page in range(2, 31):
    if page % 10 == 1: #페이지가 11, 21, 31pg이렇게 넘어가려면 다음버튼을 누르도록 설정하는 코드. 아이디어 좋은 것 같다.
        next_btn = driver.find_element(By.CSS_SELECTOR, "#contentData > div > div.movie-idv-story > nav > a.control.next") #'>'모양 버튼을 누른다(11, 21, 31pg로 가는 버튼)
        next_btn.click() #버튼을 클릭한다.
    else:
        page_btn = driver.find_element(By.LINK_TEXT, str(page)) #하이퍼링크 #두 번째 매개변수값은 가려는 페이지 번호 #원래 R에서는 CSS선택자 지정해서 썼는데 여기서는 LINK_TEXT를 쓰니까 너무 편하다고 한다. 두번째 매개변수값과 같은 <a>태그 컨텐트 내용에 해당하는 것을 찾는 것 같다.
        page_btn.click()
    #driver.implicitly_wait(3) #GET할때 영향을 주지 여기서처럼 클릭이벤트할때 쓰는 함수가 아니다. #원래 강사가 삭제하라 했는데 나둠.
    time.sleep(2) #2초 쉴 시간을 준다.
    
    points.extend([p.text for p in driver.find_elements(By.CLASS_NAME,'story-point')])
    recommends.extend([r.text for r in driver.find_elements(By.CLASS_NAME,'story-recommend')])
    txts.extend([t.text for t in driver.find_elements(By.CLASS_NAME,'story-txt')])
    
with open('output/movie.csv', "wt", encoding="utf-8") as f: #csv 파일을 저장한다. 
    f.write("score,genre,review\n") #제목행
    for p, r, t in zip(points, recommends, txts): #첫번째 엘리먼트끼리, 두번째 엘리먼트끼리, 세번째 엘리먼트끼리 묶어서 저장
        #t = '"' + t.replace('"', '""') + '"' #큰따옴표룰 앞뒤로 붙여서 넣는 거다.('"'는 작은 따옴표 사이에 큰 따옴표가 있는 모양임) #t.replace('"', '""')는 인용부호 하나짜리를 두개짜리로 바꾸는 것이다. 근데 굳이 이렇게 안 해도 된다고 한다(앞뒤로 인용부호 안 줘도 된다). 학생이 과제로 낸 코드를 그대로 가져왔다고 함.
        f.write(",".join([p, r, t])+"\n") #','로 구분해서 문자열을 만들어서 파일에 쓰고 행 단위로 저장하도록 했다.  
```

```python
#위에 소스1셀과 이어지는 셀 #위에 코드에서 검색창에 데이터입력하고 엔터키까지 쳤고 여기서는 데이터를 실제로 받아온다.

driver.switch_to.frame("searchIframe") #IFrmae 태그는 하나의 웹페이지를 하나로만 하는 게 아니라 여러개의 웹피이지로 나누는 것? 건물 하나를 여러 집으로 나눈 것과 같다. 한 페이지는 하나의 html로 구성하는건데, 윗단에는 a.html 밑단에는 b.html 또는 좌측에는 c.html 우측에는 d.html 이렇게 서로다른 웹자원을 하나의 페이지가 구성되도록 하고 싶다면 별개의 도규먼트 영역을 정의를 해서 거기에 출력되도록 해야한다. 추가 html을. 원래 하나의 도큐먼트에는 하나의 html만 렌더링할 수 있다. (사실 뭔소린지 아직 이해가 어렵다)
#메인 프레임 영역이 아닌 서브 프레임인 ifrmae 영역으로 스위치해라. iframe 태그를 포함하고 있는 영역을 스크랩핑하기 위한 것이다. # searchIFrame이라는 서브프레임에서 찾는다.
#searchIframe은 frame 태그의 id속성

while True :    #검색창을 별개의 도큐먼트로 IFrame 태그로 채웠다고 한다.(뭔소리야)원래 메인 도큐먼트 영역만 사용할 수 있게 해주지 서브 도큐먼트 영역을 사용하도록 해주지 않는데 서브 도큐먼트 영역에서 스크랩핑하는거다. IFrame 영역이 지도 검색창이라는 것 같다. IFrame이라는 태그로 지도에 포함된 것은 맞지만, 지도와 별개로 검색창은 별개의 도큐먼트라는 것 같다.
    names = driver.find_elements(By.CSS_SELECTOR, "span.place_bluelink") #도서관의 이름
    addrs = driver.find_elements(By.CSS_SELECTOR, "span.hClKF") #도서관의 주소를 받는다.
    if len(names) == 0 :
        print("추출되는 도서관이 더 이상 없슈")
        break #while True로 된 무한루프 종료
    for i in range(len(names)) : #추출된 개수만큼 반복하여 화면에 내보낸다.
        print(names[i].text,addrs[i].text, sep=", ", end="\n") #한 페이지에 원래 50~60개씩 보여주는데 10개만 출력된다. 실제로 웹페이지에서 스크롤바를 내려보면 스크롤바의 크기가 줄어들면서 더 나온다. #스타벅스 홈페이지에서 3개만 있다가 스크롤바를 내리면 더 나왔던것과 같다. 즉 스크롤바 처리를 해줘야 한다.
    print("------------------------------------------------")
    linkurl = '#app-root > div > div.XUrfU > div.zRM9F > a:nth-child(7)' #다음페이지로 옮겨가는 '>버튼'의 css선택자.
    try :
        linkNum = driver.find_element(By.CSS_SELECTOR, linkurl)
    except : #맨 끝페이지까지 가면 꺽새괄호버튼(다음페이지로 가는 버튼)이 사라지니까 무한반복문을 종료시킴
        print("더 이상 다음 페이지 없음")
        break
    linkNum.click()  #클릭이벤트 발생시킨다.
    time.sleep(5) #클릭이벤트와 관련해서 시간을 기다려주는 것으로 기억한다. 
print("스크래핑 종료")
#driver.quit() #브라우저 창을 닫는다.
```

## 아래 코드는 책 리뷰 관련 코드다

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service('C:/Temp/chromedriver'))
driver.implicitly_wait(3) 

driver.get("http://www.yes24.com/Product/goods/40936880")
import time
temp_list = []
time.sleep(3)
stopFlag = False
while True :
  for n in range(4, 13) :
    linkurl = '#infoset_reviewContentList > div.review_sort.sortBot > div.review_sortLft > div > a:nth-child('+str(n)+')'
    linkNum = driver.find_element(By.CSS_SELECTOR, linkurl)
    linkNum.click()   
    time.sleep(2)
    readLinks = driver.find_elements(By.CSS_SELECTOR, '#infoset_reviewContentList span.review_more')
    ##infoset_reviewContentList > div:nth-child(4) > div.reviewInfoBot.crop > a > div > span
    for readlink in readLinks :
      #readlink.click()
      driver.execute_script("arguments[0].click();", readlink) #arguments[0]은 readlink다
      time.sleep(3)

    reviewList = driver.find_elements(By.CSS_SELECTOR, '#infoset_reviewContentList div.reviewInfoBot.origin div.review_cont')
    time.sleep(2)

    for review in reviewList :    
      temp_list.append(review.text) #비어있는 리스트에 append처리 하고 있다. 
    
    if len(reviewList) < 5 :
      stopFlag = True
      break
      
  if stopFlag == True :
    break
  nextPage = '#infoset_reviewContentList > div.review_sort.sortBot > div.review_sortLft > div > a.bgYUI.next'
  linkNum = driver.find_element(By.CSS_SELECTOR, nextPage)
  linkNum.click()
  time.sleep(2)

wfile = open("output/yes24file.txt","w", encoding="utf-8") 
wfile.writelines(temp_list) #각각 하나의 행이 된다.
wfile.close()

driver.quit()
```

## 스타벅스 매장 관련 코드

```python
#R시간에 해본 것을 파이썬으로 작성한 코드임
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service('C:/Temp/chromedriver')) #셀레니움 객체
driver.implicitly_wait(3) 

driver.get("https://www.istarbucks.co.kr/store/store_map.do")
import time
time.sleep(2)
loca = driver.find_element(By.CLASS_NAME, 'loca_search') #원래 퀵검색이 디폴트인데 지역검색을 하도록 설정한다.
loca.click() #지역검색 클릭
time.sleep(2) #클릭하고 2초 기다려줌
f_link = driver.find_element(By.CSS_SELECTOR, "div.loca_step1_cont > ul > li:nth-child(1) > a")
f_link.click()
time.sleep(2)
s_link = driver.find_element(By.CSS_SELECTOR, "#mCSB_2_container > ul > li:nth-child(1) > a")
s_link.click()
time.sleep(2)
shopList = driver.find_elements(By.CSS_SELECTOR, "#mCSB_3_container > ul > li") #서울 매장 개수를 여기서 읽는다.(26번째 라인에서)

temp_list = []
time.sleep(3)
count = 0
total = len(shopList) #서울 매장 개수
print(f"서울에는 총 {total}개의 매장이 있어요...")
for shop in shopList :    
    count += 1
    #print(count)
    shoplat = shop.get_attribute("data-lat")
    shoplong = shop.get_attribute("data-long")
    shopname = shop.find_element(By.TAG_NAME, "strong")
    shopinfo = shop.find_element(By.TAG_NAME, "p")
    splitinfo = shopinfo.text.split('\n') #첫번째 행은 주소, 두번째행은 전화번호인데 그걸 나눈다.
    if(len(splitinfo) == 2):
        addr = splitinfo[0]
        phonenum = splitinfo[1]
    temp_list.append([shopname.text, shoplat, shoplong, addr, phonenum]) #리스트로 만들어서 append한다. 그래서 2차원 리스트가 된다.
    if count != total and count % 3 == 0:
        driver.execute_script("var su = arguments[0]; var dom=document.querySelectorAll('#mCSB_3_container > ul > li')[su]; dom.scrollIntoView();", count)
with open('output/starbucks_shop.txt', "wt", encoding="utf-8") as f:
    for item in temp_list :
        f.write(str(item)+'\n')

#driver.quit() #웹브라우저 창을 닫는다.
```
