11. card_open_rson_nm 필드는 81이면 신용카드 83이면 신용체크카드를 의미한다 
- 개설유형 필드를 새로 생성하고 해당 코드에 맞는 유형으로 데이터를 변환하시오


````python
#raw_data_open['성별'] = np.where(raw_data_open['sex_cd'] == 1, 'M', 'F') #https://nalara12200.tistory.com/162

#raw_data_open[raw_data_open['card_open_rson_nm']==81] #https://computer-science-student.tistory.com/375
#raw_data_open['card_open_rson_nm']==81

types = [] #빈리스트 생성 #https://nalara12200.tistory.com/162
for row in raw_data_open['card_open_rson_nm'] : #'card_open_rson_nm' 값을 얻어옴
    if row == 81:
        types.append('신용카드') #개설 유형이 81이면 grades에 '신용카드'를 추가
    elif row == 83:
        types.append('신용체크카드') #개설 유형이 83이면 grades에 '신용체크카드'를 추가
        
raw_data_open['개설유형'] = types #'개설유형'열에 순차대로 리스트의 원소를 대입
````

````python
raw_data_open.head()
#raw_data_open[raw_data_open['card_open_rson_nm']==83] #신용체크카드가 너무 안 나와서 card_open_rson_nm값이 83인 경우 '신용체크카드'로 제대로 적용됐는지 확인
````

14. 카드개설연도 필드에 대해서 각 연대별로 카테고리를 구분하여 개설연대 필드를 생성하시오.
- 1990년대, 2000년대, 2010년대

````python
raw_data_open['카드개설연도'] = raw_data_open['카드개설연도'].astype(int) #https://sikaleo.tistory.com/39

years = [] #빈리스트 생성 #https://nalara12200.tistory.com/162
for row in raw_data_open['카드개설연도'] : #'카드개설연도' 값을 얻어옴
    if (row - 2010) >= 0:
        years.append('2010년대') #개설 년도가 2010년도이면 years에 '2010년대'를 추가
    elif (row - 2000) >= 0:
        years.append('2000년대')
    elif 1990 <= row < 2000:
        years.append('1990년대')
        
raw_data_open['개설연대'] = years #'개설유형'열에 순차대로 리스트의 원소를 대입
raw_data_open.head()
````

15. 카드 개설 df와 차주정보 df를 memberID를 기준으로 병합하여 새로운 df로 저장하고 확인하시오
- inner_join(두 데이터셋 간의 중복된 요소만을 Join)

````python
 # http://www.gisdeveloper.co.kr/?p=8255
fin_data = pd.merge(raw_data, raw_data_open, left_on='사용자ID', right_on='사용자ID', how='inner')
fin_data.head()
````
